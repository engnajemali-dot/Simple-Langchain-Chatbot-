```python
import os
from dotenv import load_dotenv

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
```

```python
# Load environment variables
load_dotenv()
```

```python
# Used in colab
# os.environ['OPENAI_API_KEY'] = 'Your OPENAI_API_KEY'
```

```python
# Initialize model
llm = ChatOpenAI(
    model="gpt-4.1-mini",
    temperature=0.7
)
```

A system prompt is a set of instructions or guidelines given to a language model to establish its persona, behavior, or constraints for a given conversation. It helps shape the AI's responses and ensures it stays on topic or follows specific rules.

```python
import os

# Create the 'prompts' directory if it doesn't exist
if not os.path.exists('prompts'):
    os.makedirs('prompts')

# Example content for system_prompt.txt
system_prompt = (
    """You are a professional AI assistant.

      Rules:
        - Be concise
        - Answer clearly
        - Keep context from conversation"""
)

# Write the content to the file
with open('prompts/system_prompt.txt', 'w') as file:
    file.write(system_prompt)

print("Created 'prompts/system_prompt.txt' with example content.")
```

```python
# Conversation memory
chat_history = [
    SystemMessage(content=system_prompt)
]

print("\n🤖 AI Chatbot Started!")
print("Type 'exit' to quit.\n")

while True:
  # User input
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("\n👋 Goodbye!")
        break
```

```python
# Add user message to memory
chat_history.append(
        HumanMessage(content=user_input)
    )
```

```python

# Get AI response
response = llm.invoke(chat_history)

    # Print response
print(f"\nAI: {response.content}\n")
```

```python
# Save AI response to memory
chat_history.append(
    AIMessage(content=response.content)
    )
```
